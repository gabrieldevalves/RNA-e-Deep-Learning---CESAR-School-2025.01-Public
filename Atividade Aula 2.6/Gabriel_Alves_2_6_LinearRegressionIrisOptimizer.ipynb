# Regressão Linear com PyTorch

## Objetivos

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se:
- a função de perda MSE do PyTorch,
- treinamento dos parâmetros via gradiente descendente usando o otimizador.
- A rede é criada com uma camada nn.Linear()

## Importação dos pacotes

```python
%matplotlib inline
import torch
from torch import nn, optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
```

## Criação do modelo da rede

```python
model = torch.nn.Linear(2, 1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight.data
```

### Testando o predict da rede

```python
model(Variable(torch.ones((5,2))))
```

## Treinamento

### Definindo função de perda e otimizador

```python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.3)
```

### Laço de treinamento

```python
num_epochs = 120
w0_list = []
w1_list = []
for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    # forward - predict
    out = model(inputs)

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    #loss cálculo da função de perda
    loss = criterion(out, target)

    # backward e otimizador
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(x_train_bias)
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

   função `criterion` utilizada no laço de treinamento.

```python
y_pred = model(x_train_bias)
loss = criterion(y_pred, y_train)
print(f'MSE depois do treinamento: {loss.item():.6f}')
```

- 2) Faça igual o exercício do notebook anterior, de plotar um gráfico scatterplot para
   mostrar a evolução dos parâmetros durante o treinamento pelo gradiente descendente.

```python
plt.figure(figsize=(8, 6))
plt.scatter(w0_list, w1_list, c='blue', marker='o', label='Evolução do Parâmetro')
w0_old = None
w1_old = None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old is not None:
        plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
                  head_length=0.005, head_width=0.005, shape='full',
                  length_includes_head=True, color='red', alpha=0.7)
    w0_old, w1_old = w0, w1

plt.xlabel('w0 (Bias)')
plt.ylabel('w1 (Peso)')
plt.title('Evolução dos parâmetros durante o treinamento')
plt.grid(True)
plt.legend()
plt.show()
```

# Aprendizados

Cálculo do MSE Pós-Treinamento: Calculamos o Erro Quadrático Médio (MSE) após o treinamento da rede neural. Este valor, que foi de aproximadamente 0.021181, nos informa o quão bem nosso modelo se ajustou aos dados de treinamento. Um MSE baixo, como este, indica que o modelo fez boas previsões, com pouca diferença entre os valores previstos e os reais.

Evolução dos Parâmetros (Gradiente Descendente): Ao plotar a evolução dos parâmetros w0 (bias) e w1 (peso), visualizamos como o algoritmo de Gradiente Descendente ajusta esses valores a cada época de treinamento. O gráfico mostrou o "caminho" que os parâmetros percorreram no espaço de pesos, partindo de seus valores iniciais e convergindo gradualmente para um ponto onde a função de perda (MSE) é minimizada. Isso demonstra o funcionamento iterativo da otimização de modelos de aprendizado de máquina.
